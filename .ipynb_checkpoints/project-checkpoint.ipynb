**filtering dataframe without screen_view,user_engagement and session_start**

```python
df1 = df[(df.event_name != 'screen_view') & (df.event_name != 'user_engagement')  & (df.event_name != 'session_start') ]
```

**new column in_app_purchase based on events in_app_purchase**

```python
df1['in_app_purchase'] = np.where(df1['user_pseudo_id'].isin(df1[df1.event_name=='in_app_purchase']['user_pseudo_id'].values),1,0)
```

**filtering only in_app_purchase users**

```python
df2 = df1[df1.in_app_purchase==1]
```

```python
df2 = df2[df2.user_pseudo_id != '652EA044B9BE4E3FA3FA8900BBB0C729'] 
#there are 2 app purchase for this id thereby this was excluded
```
